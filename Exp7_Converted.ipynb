```
def largest_of_three(a,b,c): return max(a,b,c)
print(largest_of_three(12,45,30))
```

```
def factorial(n):
    r=1
    for i in range(2,n+1): r*=i
    return r
print(factorial(6))
```

```
def is_prime(n):
    if n<=1:return False
    for i in range(2,int(n**0.5)+1):
        if n%i==0:return False
    return True
print(is_prime(29))
```

```
def count_vowels(s): return sum(1 for ch in s.lower() if ch in 'aeiou')
print(count_vowels('Prerana'))
```

```
def sum_even(lst): return sum(x for x in lst if x%2==0)
print(sum_even([1,2,3,4,5,6]))
```

```
def is_palindrome(s):
    t=''.join(ch.lower() for ch in s if ch.isalnum())
    return t==t[::-1]
print(is_palindrome('Level'))
```

```
def fact_rec(n):
    if n<=1:return 1
    return n*fact_rec(n-1)
print(fact_rec(5))
```

```
def fib(n):
    if n<=0:return []
    if n==1:return [0]
    if n==2:return [0,1]
    seq=fib(n-1)
    seq.append(seq[-1]+seq[-2])
    return seq
print(fib(7))
```

```
def sum_n(n):
    if n==0:return 0
    return n+sum_n(n-1)
print(sum_n(10))
```

```
def rev_rec(s):
    if s=='':return ''
    return rev_rec(s[1:])+s[0]
print(rev_rec('sagar'))
```

```
def count_digits(n):
    n=abs(n)
    if n<10:return 1
    return 1+count_digits(n//10)
print(count_digits(123456))
```

```
def power(x,y):
    if y==0:return 1
    return x*power(x,y-1)
print(power(2,8))
```

```
def gcd(a,b):
    if b==0:return a
    return gcd(b,a%b)
print(gcd(48,18))
```

```
def is_palindrome_num(n):
    s=str(n)
    return s==s[::-1]
print(is_palindrome_num(1221))
```

```
def sum_list(lst):
    if not lst:return 0
    return lst[0]+sum_list(lst[1:])
print(sum_list([1,2,3,4,5]))
```

```
def print_reverse(lst):
    if not lst:return []
    return print_reverse(lst[1:])+[lst[0]]
print(print_reverse([1,2,3,4]))
```

```
def len_rec(s):
    if s=='':return 0
    return 1+len_rec(s[1:])
print(len_rec('sachin'))
```

```
def dec_to_bin(n):
    if n==0:return '0'
    if n==1:return '1'
    return dec_to_bin(n//2)+str(n%2)
print(dec_to_bin(13))
```

```
def min_rec(lst):
    if len(lst)==1:return lst[0]
    m=min_rec(lst[1:])
    return lst[0] if lst[0]<m else m
print(min_rec([5,3,9,1,7]))
```

```
def sum_digits(n):
    n=abs(n)
    if n<10:return n
    return n%10+sum_digits(n//10)
print(sum_digits(12345))
```

```
def is_pal_rec(s):
    s=''.join(ch.lower() for ch in s if ch.isalnum())
    if len(s)<=1:return True
    if s[0]!=s[-1]:return False
    return is_pal_rec(s[1:-1])
print(is_pal_rec('Madam'))
```
