```
first = "Avnish"
last = "Deshpande"
nick = first[:3] + last[:2]
full = first + " " + last
search_letter = 'a'
found = search_letter.lower() in full.lower()
print(nick, found)
```

```
s = "PythonProgrammingBasics"
part1 = s[:6]
part2 = s[6:16]
part3 = s[16:]
joined = part1 + "-" + part3
print(part1, part2, part3, joined)
print("Contains 'gram'?", "gram" in s)
```

```
city = "Pune"; state = "Maharashtra"
code = city[:2].upper() + state[:2].upper()
search_char = 'u'
found = search_char.lower() in city.lower()
print(code, found)
```

```
email = "rahul.sharma23@gmail.com"
username = email.split('@')[0]
new_domain = "college.edu"
new_email = username + "@" + new_domain
print(new_email, any(ch.isdigit() for ch in username))
```

```
word = "radha"
rev = word[::-1]
concat = word + rev
print(rev)
print(concat)
print(next((ch for ch in word if ch in rev), None))
```

```
text = "Knowledge is power"
parts = text.split()
new = parts[2] + " is " + parts[0]
print(new)
print("Contains 'power'?", "power" in text)
```

```
w1, w2 = "learning", "coding"
phrase = w1[:4] + w2[-4:]
print(phrase)
print("Contains 'ing'?", "ing" in phrase)
```

```
s1 = "Data Science is fun."; s2 = "Machine Learning is powerful."
part1 = s1.split()[0] + " " + s1.split()[1]
part2 = "Learning"
print(part1 + " & " + part2)
print("Learning" in s1 or "Learning" in s2)
```

```
movie = "3 Idiots"; year = 2009
code = ''.join(ch for ch in movie if ch.isalnum())[:3].upper() + str(year)[-2:]
print(code)
print("Idiot" in movie)
```

```
name = "Priya"; birth = 1998
pwd = name[:2].lower() + str(birth)[-2:] + name[-1].upper()
print(pwd)
print(any(ch.isdigit() for ch in pwd))
```

```
college = "COEP Pune"; dept = "Computer Engineering"
code = ''.join([w[0] for w in dept.split()[:2]]).upper() + college.split()[0][:2].upper()
print(code)
print("Tech" in college or "Tech" in dept)
```
